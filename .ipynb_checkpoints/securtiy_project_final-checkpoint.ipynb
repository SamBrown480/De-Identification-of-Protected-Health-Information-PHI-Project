## Table 3
Supressed Patient Values

```python
suppressed_df = df.copy()
suppressed_df['id'] = 0
def random_pop(df, column):
  for i in df.index:
    random_num = random.random()
    if random_num < .4:
      df.loc[i, col] = np.nan 


pop_columns = ['age','bmi','charges','medicines','illness_history','country','marital_status']
for col in pop_columns:
  random_pop(suppressed_df, col)

```

```python
suppressed_df.head(6)
```

```python
# 1. Count the number of suppressed values in each column
suppressed_counts = suppressed_df.isna().sum()

# 2. Check for unique combinations
unique_combinations = suppressed_df.drop(columns=['id', 'sex', 'smoker', 'alcohol', 'allergies']).dropna().drop_duplicates().shape[0]

# 3. K-anonymity evaluation
quasi_identifiers = ['age', 'bmi', 'marital_status', 'country', 'medicines', 'illness_history', 'charges']
k_anonymity = suppressed_df.groupby(quasi_identifiers).size().reset_index(name='count')
min_k_anonymity = k_anonymity['count'].min()

suppressed_counts, unique_combinations, min_k_anonymity
# print(suppressed_counts)
```

```python
# Calculate the total number of records
total_records = len(df)

# Calculate the percentages of suppressed values
suppressed_percentages = (suppressed_counts / total_records) * 100

# Format the output
formatted_suppressed_percentages = suppressed_percentages.to_frame(name='Suppressed Percentage').reset_index()
formatted_suppressed_percentages.columns = ['Attribute', 'Suppressed Percentage']

# Unique combinations percentage
unique_combinations_percentage = (unique_combinations / total_records) * 100

# Format the k-anonymity output
k_anonymity_output = {
    'Minimum K-Anonymity': min_k_anonymity,
    'Unique Combinations Percentage': unique_combinations_percentage
}

formatted_suppressed_percentages, k_anonymity_output
```

```python
gen_df = df.copy()
# for age lets try groups of 10 years
age_groups = ['1-10 years old', '10-20 years old', '20-30 years old', '30-40 years old',
              '40-50 years old','50-60 years old','60-70 years old', '70-80 years old',
              '80-90 years old','90-100 years old', '100+ years old']
hist, bins = np.histogram(gen_df['age'], bins=10)
gen_df['age'] = pd.cut(gen_df['age'], bins=bins, labels=age_groups[:-1])
#bmi will be by 3
bmi_groups = ['15-18 bmi','18-21 bmi','21-24 bmi','24-27 bmi',
              '27-30 bmi','30-33 bmi','33-36 bmi','36-39 bmi','39-41 bmi',
              '41-44 bmi','44+ bmi']
hist, bins = np.histogram(gen_df['bmi'], bins=10)
gen_df['bmi'] = pd.cut(gen_df['bmi'], bins=bins, labels=bmi_groups[:-1])
gen_df.head(5)
#charges
charges_groups = ['~120 Dollars', '~130 Dollars', '~140 Dollars', '~150 Dollars',
                  '~160 Dollars', '~170 Dollars','~180 Dollars',
                  '~190 Dollars', '~200 Dollars']
hist, bins = np.histogram(gen_df['charges'], bins=8)
gen_df['charges'] = pd.cut(gen_df['charges'], bins=bins, labels=charges_groups[:-1])
gen_df.head(5)
```

```python
df.head(5)
```

```python
# Check for unique combinations
unique_combinations_gen = gen_df.drop(columns=['id', 'sex', 'smoker', 'alcohol', 'allergies']).dropna().drop_duplicates().shape[0]

# K-anonymity evaluation
quasi_identifiers_gen = ['age', 'bmi', 'marital_status', 'medicines', 'illness_history', 'charges']
k_anonymity_gen = gen_df.groupby(quasi_identifiers_gen).size().reset_index(name='count')
min_k_anonymity_gen = k_anonymity_gen['count'].min()

print(f'Unique Combinations after Generalization: {unique_combinations_gen}')
print(f'Minimum K-Anonymity: {min_k_anonymity_gen}')
```

##Table 5
Randomized patient values.

```python
df_randomizedValues = df.copy()
rando_3 = random.randint(-3,3)
#age +/- 2
for index, value in df_randomizedValues['age'].items():
    rando_2 = random.randint(-2,2)
    df_randomizedValues.at[index,'age'] = value + (rando_2)


#bmi +/- 3
for index, value in df_randomizedValues['bmi'].items():
    rando_3 = random.randint(-3,3)
    df_randomizedValues.at[index,'bmi'] = value + (rando_3)

#charges +/5.00
for index, value in df_randomizedValues['charges'].items():
    rando_float = random.uniform(-5.00,5.00)
    new_value = value + (rando_float)
    df_randomizedValues.at[index,'charges'] = round(new_value, 2)
df_randomizedValues.head(5)
```

```python
df.head(5)
```

```python
# randomized

original_stats = df[['age', 'bmi', 'charges']].agg(['mean', 'var'])
randomized_stats = df_randomizedValues[['age', 'bmi', 'charges']].agg(['mean', 'var'])

#Step 2: Check for Unique Combinations
unique_combinations_random = df_randomizedValues.drop(columns=['id', 'sex', 'smoker', 'alcohol', 'allergies']).dropna().drop_duplicates().shape[0]

##Step 3: K-Anonymity Evaluation
quasi_identifiers_random = ['age', 'bmi', 'marital_status', 'country', 'medicines', 'illness_history', 'charges']
k_anonymity_random = df_randomizedValues.groupby(quasi_identifiers_random).size().reset_index(name='count')
min_k_anonymity_random = k_anonymity_random['count'].min()

#Display the results
print("Original Statistics:\n", original_stats)
print("\nRandomized Statistics:\n", randomized_stats)
print(f"\nUnique Combinations after Randomization: {unique_combinations_random}")
print(f"Minimum K-Anonymity after Randomization: {min_k_anonymity_random}")
```

```python
df_k_anonymize.groupby(['age','bmi','charges']).size().reset_index(name='count')
```

Without Generalizations we dont have any 'groups' that have more than one value in the data set. So lets generalize.

```python
df_k_anonymize['age_group'] = pd.cut(df_k_anonymize['age'], bins=3)
df_k_anonymize['bmi_group'] = pd.cut(df_k_anonymize['bmi'], bins=3)
df_k_anonymize['charges_group'] = pd.cut(df_k_anonymize['charges'], bins=2)

df_grouped = df_k_anonymize.groupby(['age_group','bmi_group',
                                     'charges_group']).size().reset_index(name='count')
df_grouped[df_grouped['count'] <= 12]
```

Now that we have no group that has only one value in it we can say that these columns are k-anonimyzed where k = 2.

Now since the data that could be used to identify a person is k-anonmiyzed we can make a 'distrubated' data set with health information. So lets say some company wanted to train a model that needs data on patients that are smokers and how much their visit costed. Lets give them a k-anonimized data set with those fields.

```python
df_grouped_smoker = df_k_anonymize.groupby(['age_group','bmi_group',
                                            'charges_group','smoker']).size(
                                            ).reset_index(name='count')
```
